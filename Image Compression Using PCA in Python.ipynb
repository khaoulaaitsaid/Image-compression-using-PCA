```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
plt.imshow(img)
```

```python
r, g, b = cv2.split(img)
r, g, b = r/255, g/255, b/255
```

```python
plt.imshow(r)
```

```python
plt.imshow(g)
```

```python
plt.imshow(b)
```

```python
pca_components = 50
pca_r = PCA(n_components=pca_components)
reduced_r = pca_r.fit_transform(r)
```

```python
pca_components = 50

pca_r = PCA(n_components=pca_components)
pca_g = PCA(n_components=pca_components)
pca_b = PCA(n_components=pca_components)


reduced_r = pca_r.fit_transform(r)  
reduced_g = pca_g.fit_transform(g)  
reduced_b = pca_b.fit_transform(b)

reconstructed_r = pca_r.inverse_transform(reduced_r)
reconstructed_g = pca_g.inverse_transform(reduced_g)
reconstructed_b = pca_b.inverse_transform(reduced_b)
```

```python
reconstructed_channels = [reconstructed_r, reconstructed_g, reconstructed_b]
img_reconstructed = cv2.merge(reconstructed_channels)
plt.imshow(img_reconstructed)
```

```python
img = cv2.cvtColor(cv2.imread('pi.png'), cv2.COLOR_BGR2RGB)
r, g, b = cv2.split(img)
r, g, b = r/255, g/255, b/255
pca_components = 50

pca_r = PCA(n_components=pca_components)
pca_g = PCA(n_components=pca_components)
pca_b = PCA(n_components=pca_components)


reduced_r = pca_r.fit_transform(r)  # Remplacez 'r' par vos donnes de composants rouges
reduced_g = pca_g.fit_transform(g)  # Remplacez 'g' par vos donnes de composants verts
reduced_b = pca_b.fit_transform(b)

reconstructed_r = pca_r.inverse_transform(reduced_r)
reconstructed_g = pca_g.inverse_transform(reduced_g)
reconstructed_b = pca_b.inverse_transform(reduced_b)
reconstructed_channels = [reconstructed_r, reconstructed_g, reconstructed_b]
img_reconstructed = cv2.merge(reconstructed_channels)
plt.imshow(img_reconstructed)
```
